## Part 1: What is Pandas?

NumPy is great for arrays of numbers. But real economic datasets have:
- **Mixed types** — country names (strings) alongside GDP (floats) and year (int)
- **Column labels** — variables have names, not just positions
- **Missing values** — not every country reports every indicator
- **Row labels** — observations have identifiers

**Pandas** handles all of this through its core object: the **DataFrame** — essentially a spreadsheet inside Python.

```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
```

## Part 2: Creating a DataFrame

The most common way: from a Python dictionary where **keys are column names** and **values are lists**.

```python
# Build our 12-country development dataset as a DataFrame
data = {
    'country'       : ['USA', 'France', 'Japan', 'UAE',
                       'Brazil', 'Mexico', 'South Korea',
                       'Vietnam', 'Nigeria', 'Ghana',
                       'Ethiopia', 'Kenya'],

    'region'        : ['North America', 'Europe', 'East Asia', 'Middle East',
                       'Latin America', 'Latin America', 'East Asia',
                       'East Asia', 'Sub-Saharan Africa', 'Sub-Saharan Africa',
                       'Sub-Saharan Africa', 'Sub-Saharan Africa'],

    'income_group'  : ['High', 'High', 'High', 'High',
                       'Upper-Middle', 'Upper-Middle', 'High',
                       'Lower-Middle', 'Lower-Middle', 'Lower-Middle',
                       'Low', 'Low'],

    'gdp_per_capita': [76399.0, 40886.0, 33815.0, 49532.0,
                        9000.5,  9926.0, 32255.0,
                        4163.5,  2065.8,  2363.0,
                         925.1,  1838.0],

    'population'    : [333271341,  67897000, 125124989,   9890400,
                       215313498, 127504125,  51628117,
                        97468029, 218541000,  32395450,
                       123380000,  54027487],

    'poverty_rate'  : [1.2, 0.1, 0.7, 0.0,
                       5.8, 7.3, 0.2,
                       1.5, 39.1, 24.8,
                       26.1, 36.1],

    'growth_rate'   : [2.0, 2.5, 1.5, 3.5,
                       3.0, 2.5, 2.7,
                       6.1, 3.2, 3.8,
                       7.0, 5.2],

    'year'          : [2022] * 12
}

df = pd.DataFrame(data)
df
```

Notice: Pandas displays this as a clean table — row indices on the left, column names on top. This is a **DataFrame**.

## Part 3: Inspecting a Dataset

The first thing you always do with a new dataset: **look at it carefully before touching it.**

```python
# Shape: (rows, columns)
print("Shape:", df.shape)
print(f"  → {df.shape[0]} countries, {df.shape[1]} variables")
```

```python
# Column names and data types
print(df.dtypes)
```

```python
# First and last few rows
print("--- First 4 rows ---")
display(df.head(4))

print("--- Last 4 rows ---")
display(df.tail(4))
```

```python
# Statistical summary of all numeric columns
df.describe().round(2)
```

```python
# Missing values — critical check before any analysis
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nTotal missing: {df.isnull().sum().sum()}")
```

```python
# Now let's introduce a missing value to see how it looks
df_with_missing = df.copy()
df_with_missing.loc[8, 'poverty_rate'] = np.nan   # Nigeria's poverty rate is missing
df_with_missing.loc[10, 'growth_rate'] = np.nan   # Ethiopia's growth rate is missing

print("Missing values after introducing NaNs:")
print(df_with_missing.isnull().sum())
```

```python
# Unique values in categorical columns
print("Income groups:", df['income_group'].unique())
print("Regions:",       df['region'].unique())
print()
print("Countries per income group:")
print(df['income_group'].value_counts())
```

## Part 4: Loading Real Data from the World Bank

In practice, we load data from files or APIs — not type it by hand.  
Let's pull real World Bank data using the `wbgapi` library.

```python
# Install World Bank API (only needed once in Colab)
!pip install wbgapi -q
```

```python
import wbgapi as wb

# Download GDP per capita (current USD) for all countries, 2000-2022
# Indicator: NY.GDP.PCAP.CD
gdp_wb = wb.data.DataFrame('NY.GDP.PCAP.CD',
                             time=range(2000, 2023),
                             labels=True)

# Clean up column names
gdp_wb.columns = [str(c).replace('YR', '') for c in gdp_wb.columns]
gdp_wb = gdp_wb.reset_index()

print("Shape:", gdp_wb.shape)
gdp_wb.head()
```

```python
# Filter to our 12 countries
our_countries = ['United States', 'France', 'Japan', 'United Arab Emirates',
                 'Brazil', 'Mexico', 'Korea, Rep.',
                 'Vietnam', 'Nigeria', 'Ghana',
                 'Ethiopia', 'Kenya']

gdp_panel = gdp_wb[gdp_wb['Country'].isin(our_countries)].copy()
gdp_panel = gdp_panel.reset_index(drop=True)

print("Countries found:", gdp_panel['Country'].nunique())
gdp_panel[['Country', '2000', '2010', '2022']].round(0)
```

## Part 5: Selecting Columns and Rows

```python
# Selecting a single column — returns a Series
gdp_col = df['gdp_per_capita']
print(type(gdp_col))
print(gdp_col)
```

```python
# Selecting multiple columns — returns a DataFrame
subset = df[['country', 'gdp_per_capita', 'poverty_rate']]
subset
```

```python
# .loc — select by LABEL (row index label, column name)
print("Row 0 (USA):")
print(df.loc[0])
print()
print("USA's GDP per capita:")
print(df.loc[0, 'gdp_per_capita'])
```

```python
# .iloc — select by POSITION (integer index)
print("First 3 rows, first 3 columns:")
df.iloc[0:3, 0:3]
```

```python
# Set country as the index for more intuitive access
df_indexed = df.set_index('country')

# Now access by country name directly
print("Nigeria's data:")
print(df_indexed.loc['Nigeria'])
print()
print("Nigeria's poverty rate:", df_indexed.loc['Nigeria', 'poverty_rate'])
```

## Part 6: Filtering Rows

Filtering in Pandas works exactly like NumPy boolean indexing — but with column names.

```python
# Countries with poverty rate above 20%
high_poverty = df[df['poverty_rate'] > 20]
high_poverty[['country', 'region', 'poverty_rate', 'gdp_per_capita']]
```

```python
# Low-income countries only
low_income = df[df['income_group'] == 'Low']
low_income[['country', 'gdp_per_capita', 'poverty_rate', 'growth_rate']]
```

```python
# Combined: Sub-Saharan Africa AND GDP per capita above $1,500
ssa_above_threshold = df[
    (df['region'] == 'Sub-Saharan Africa') &
    (df['gdp_per_capita'] > 1500)
]
ssa_above_threshold[['country', 'gdp_per_capita', 'poverty_rate']]
```

```python
# .isin() — filter by a list of values
east_asia = df[df['region'].isin(['East Asia'])]
east_asia[['country', 'gdp_per_capita', 'growth_rate']]
```

```python
# .query() — SQL-style filtering, very readable
result = df.query("poverty_rate < 5 and gdp_per_capita < 50000")
result[['country', 'income_group', 'gdp_per_capita', 'poverty_rate']]
```

---
## Part 7: Sorting and Ranking

```python
# Sort by GDP per capita — richest to poorest
df.sort_values('gdp_per_capita', ascending=False)[['country', 'gdp_per_capita', 'income_group']]
```

```python
# Sort by poverty rate — worst to best
df.sort_values('poverty_rate', ascending=False)[['country', 'poverty_rate', 'region']]
```

```python
# Multi-column sort: first by region, then by GDP within region
df.sort_values(['region', 'gdp_per_capita'], ascending=[True, False])[['country', 'region', 'gdp_per_capita']]
```

```python
# Add a rank column
df['gdp_rank']     = df['gdp_per_capita'].rank(ascending=False).astype(int)
df['poverty_rank'] = df['poverty_rate'].rank(ascending=True).astype(int)  # lower poverty = better rank

df[['country', 'gdp_per_capita', 'gdp_rank', 'poverty_rate', 'poverty_rank']].sort_values('gdp_rank')
```

---
## Part 8: Adding and Modifying Columns

```python
# Total GDP = GDP per capita × population
df['gdp_total_bn'] = (df['gdp_per_capita'] * df['population'] / 1e9).round(1)

# Log GDP per capita — standard in development economics regressions
df['log_gdp'] = np.log(df['gdp_per_capita']).round(3)

# Projected GDP per capita in 10 years
df['gdp_2032'] = (df['gdp_per_capita'] * (1 + df['growth_rate']/100) ** 10).round(0)

df[['country', 'gdp_per_capita', 'gdp_total_bn', 'log_gdp', 'gdp_2032']]
```

```python
# Categorical column: classify countries by poverty severity
def poverty_category(rate):
    if rate < 2:
        return 'Very Low'
    elif rate < 10:
        return 'Low'
    elif rate < 25:
        return 'Moderate'
    else:
        return 'High'

df['poverty_category'] = df['poverty_rate'].apply(poverty_category)
df[['country', 'poverty_rate', 'poverty_category']]
```

```python
# np.where — vectorized if-else
df['fast_grower'] = np.where(df['growth_rate'] >= 5, 'Yes', 'No')
df[['country', 'growth_rate', 'fast_grower']]
```

## 📌 Summary

| Task | Method |
|------|--------|
| Create DataFrame | `pd.DataFrame(dict)` |
| Inspect | `.shape`, `.dtypes`, `.describe()`, `.isnull()` |
| Select column | `df['col']` or `df[['col1','col2']]` |
| Select by label | `df.loc[row, col]` |
| Select by position | `df.iloc[row, col]` |
| Filter rows | `df[df['col'] > value]` |
| SQL-style filter | `df.query("col > value")` |
| Sort | `df.sort_values('col')` |
| Add column | `df['new'] = ...` |
| Apply function | `df['col'].apply(func)` |
| Save | `df.to_csv('file.csv')` |
| Load | `pd.read_csv('file.csv')` |
