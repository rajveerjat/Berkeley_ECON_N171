## Part 3: Arithmetic

| Operator | Symbol | Use |
|----------|--------|-----|
| Addition | `+` | Sum of incomes |
| Subtraction | `-` | Poverty gap |
| Multiplication | `*` | Total cost |
| Division | `/` | GDP per capita |
| Exponentiation | `**` | Compound growth |

```python
# GDP per capita from totals
gdp_total_mexico  = 1322600000000   # USD (2022)
population_mexico = 127504125

gdp_pc_mexico = gdp_total_mexico / population_mexico
print(f"Mexico GDP per capita: ${gdp_pc_mexico:,.2f} USD")
```

```python
# How rich-poor is the world?
gdp_pc_usa      = 76399.0    # USD
gdp_pc_ethiopia =   925.07   # USD

ratio = gdp_pc_usa / gdp_pc_ethiopia
print(f"A US resident earns {ratio:.1f}x more than an Ethiopian resident per year.")
```

```python
# Compound growth: Vietnam growing at 6.1% per year
gdp_pc_vietnam_2022 = 4163.45
growth_rate         = 0.061
years               = 10

gdp_pc_vietnam_2032 = gdp_pc_vietnam_2022 * (1 + growth_rate) ** years
print(f"Vietnam GDP per capita in 2032 (projected): ${gdp_pc_vietnam_2032:,.2f} USD")
```

```python
# Poverty gap: how far below the poverty line is the average poor person?
poverty_line       = 2.15   # World Bank extreme poverty line (USD/day)
avg_poor_income    = 1.40   # Average income of a poor person (USD/day)

daily_gap  = poverty_line - avg_poor_income
annual_gap = daily_gap * 365

print(f"Daily poverty gap:  ${daily_gap:.2f} USD per person")
print(f"Annual poverty gap: ${annual_gap:.2f} USD per person")
```

## Part 5: Lists

A **list** stores multiple values in order — perfect for cross-country data.

```python
# A global sample — rich and poor, every continent
countries = ["USA", "France", "Japan", "UAE",
             "Brazil", "Mexico", "South Korea",
             "Vietnam", "Nigeria", "Ghana",
             "Ethiopia", "Kenya"]

print("Countries:", countries)
print("Total:", len(countries))
```

```python
# GDP per capita (USD, 2022, World Bank) — same order as countries
gdp_per_capita = [76399.0, 40886.0, 33815.0, 49532.0,
                   9000.5,  9926.0, 32255.0,
                   4163.5,  2065.8,  2363.0,
                    925.1,  1838.0]

# 0-based indexing
print("First country:",     countries[0])          # USA
print("Last country:",      countries[-1])          # Kenya
print("Vietnam GDP/capita:", gdp_per_capita[6+1])  # index 7
```

```python
# Slicing
print("High-income group:",   countries[0:4])
print("Low-income group:",    countries[-2:])
```

```python
# Summary statistics
print(f"Richest country GDP/capita:  ${max(gdp_per_capita):,.2f} USD")
print(f"Poorest country GDP/capita:  ${min(gdp_per_capita):,.2f} USD")
print(f"World average GDP/capita:    ${sum(gdp_per_capita)/len(gdp_per_capita):,.2f} USD")
print(f"Rich-poor ratio:              {max(gdp_per_capita)/min(gdp_per_capita):.1f}x")
```
