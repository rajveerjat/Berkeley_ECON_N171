## Part 1: Why NumPy?

In Lecture 1, we stored data in Python **lists**. Lists work, but they are slow and clunky for math.  
**NumPy** (Numerical Python) gives us **arrays** — fast, compact, and built for computation.

```python
import numpy as np
print("NumPy version:", np.__version__)
```

```python
# The problem with plain lists: you cannot do math directly
gdp_list = [76399.0, 40886.0, 33815.0, 49532.0,
             9000.5,  9926.0, 32255.0,
             4163.5,  2065.8,  2363.0,
              925.1,  1838.0]

# Try to multiply every element by 1000 — this FAILS with a list
try:
    result = gdp_list * 1000   # what do you think this does?
    print(result[:5])          # not what we want!
except:
    print("Error: can't do math directly on a list")
```

```python
# NumPy array: math just works
gdp = np.array(gdp_list)

# Convert GDP per capita from USD to thousands of USD
gdp_thousands = gdp / 1000
print("GDP per capita (thousands USD):")
print(gdp_thousands.round(2))
```

```python
# Speed comparison: NumPy vs Python list
import time

big_list  = list(range(1000000))
big_array = np.array(big_list)

# List: multiply each element by 2 using a loop
start = time.time()
result_list = [x * 2 for x in big_list]
list_time = time.time() - start

# NumPy: multiply entire array by 2
start = time.time()
result_array = big_array * 2
numpy_time = time.time() - start

print(f"List time:  {list_time:.4f} seconds")
print(f"NumPy time: {numpy_time:.4f} seconds")
print(f"NumPy is {list_time/numpy_time:.1f}x faster")
```

## Part 3: Vectorized Arithmetic — The Key Insight

NumPy applies operations to **every element at once** — no loops needed.  
This is called **vectorization** and is the foundation of fast data analysis.

```python
# Total GDP = GDP per capita × Population
# NumPy multiplies element-by-element automatically
gdp_total = gdp * pop

print("Total GDP (USD):")
for i, c in enumerate(countries):
    print(f"  {c:15s}: ${gdp_total[i]/1e12:,.2f} trillion")
```

```python
# Compound growth: project every country's GDP 10 years forward
growth_rates = np.array([0.020, 0.025, 0.015, 0.035,
                          0.030, 0.025, 0.027,
                          0.061, 0.032, 0.038,
                          0.070, 0.052])

gdp_2032 = gdp * (1 + growth_rates) ** 10

print("Projected GDP per capita in 2032:")
for i, c in enumerate(countries):
    print(f"  {c:15s}: ${gdp[i]:,.0f} → ${gdp_2032[i]:,.0f} USD")
```

```python
# Log transformation — widely used in development economics
# Why? GDP distributions are highly skewed; log makes them comparable
log_gdp = np.log(gdp)

print("GDP per capita vs Log GDP per capita:")
for i, c in enumerate(countries):
    print(f"  {c:15s}: ${gdp[i]:>8,.0f}  →  log = {log_gdp[i]:.2f}")
```

## Part 6: Linear Algebra — The Backbone of Econometrics

Every regression you will run in this course uses linear algebra under the hood.  
Let's get comfortable with the basics.

```python
# Dot product: weighted average GDP
# Suppose we want to compute world average GDP weighted by population share

pop_shares = pop / np.sum(pop)    # each country's share of total population
weighted_gdp = np.dot(gdp, pop_shares)

print("Population shares:")
for i, c in enumerate(countries):
    print(f"  {c:15s}: {pop_shares[i]*100:.1f}%")
print()
print(f"Population-weighted world GDP per capita: ${weighted_gdp:,.2f} USD")
print(f"Simple average GDP per capita:            ${np.mean(gdp):,.2f} USD")
```

```python
# Matrix multiplication: the heart of OLS regression
# OLS formula: beta = (X'X)^{-1} X'y
# Let's build a tiny example: does log GDP predict poverty?

# X matrix: intercept column + log GDP
ones_col = np.ones((len(gdp), 1))
log_gdp_col = np.log(gdp).reshape(-1, 1)
X = np.hstack([ones_col, log_gdp_col])   # 12 x 2 matrix

# y vector: poverty rates
y = poverty_rates

print("X matrix (first 4 rows):")
print(X[:4].round(3))
print("\ny vector (poverty rates):")
print(y)
```

```python
# OLS estimator: beta = (X'X)^{-1} X'y
XtX     = X.T @ X                          # X transpose times X
XtX_inv = np.linalg.inv(XtX)              # inverse
Xty     = X.T @ y                          # X transpose times y
beta    = XtX_inv @ Xty                    # OLS coefficients

print("OLS Coefficients:")
print(f"  Intercept (beta_0): {beta[0]:.4f}")
print(f"  Log GDP   (beta_1): {beta[1]:.4f}")
print()
print("Interpretation: a 1% increase in GDP per capita is associated with")
print(f"a {beta[1]/100:.4f} percentage point change in the poverty rate.")
```
